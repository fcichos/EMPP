```python
import numpy as np

with open('a.txt', 'r') as file_1, open('b.txt', 'r') as file_2, open('result.txt', 'w') as result:
    for a, b in zip(file_1, file_2):
        result.write(f"{np.gcd(int(a),int(b))}\n")
```

Assign the 100 numbers between -10 and 10 including -10 and 10 to the variable x using the linspace function of numpy. Assign the function values \( x²-20 \) to the variable y. Store the paired values in the file named "data.txt" in the form "x, y" (e.g. -10, 80) for each line of the file.

**Problem 3**

Write a Python function projectile_motion(v0, theta, dt) that simulates the projectile motion of an object launched at an initial velocity v0 (in m/s) and angle theta (in degrees). The function should return a tuple containing:

The maximum height reached by the object (in meters).
The total time of flight (in seconds).
The horizontal range (in meters).

Use a time step dt (in seconds) for the simulation. Advance the time inside the function with a loop. Note that calling the function is not required for the task! Can be done only with the math module. 

What if one of the arguments is 0?

Aufgabenstellung klarer!

```python
import math

def projectile_motion(v0, theta, dt):
    if(v0 == 0 or theta == 0 or dt == 0): return (0,0,0)
    
    theta_rad = math.radians(theta)
    
    v0x = v0 * math.cos(theta_rad)
    v0y = v0 * math.sin(theta_rad)
    
    x_points = []
    y_points = []
    
    x = 0
    y = 0
    t = 0
    
    while y >= 0:
        x = v0x * t
        y = v0y * t - 0.5 * 9.81 * t**2
        x_points.append(x)
        y_points.append(y)
        t += dt
    
    max_height = max(y_points)
    total_time = t - dt
    horizontal_range = x_points[-1]
    
    return (max_height, total_time, horizontal_range)
```
